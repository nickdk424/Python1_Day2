#### 2. How much water runs off a roof in a rain storm?
To calculate the runoff from any given rainfall: Take the dimensions of the footprint of the roof and convert them to inches. (So, a 50' x 20' roof is 600" x 240"). Multiply the roof dimensions by the number of inches of rainfall. As an example, 600" x 240" x 1" = 144,000 cubic inches of water for an inch of rainfall. Divide that result by 231 to get the number of gallons (because 1 gallon = 231 cubic inches). (144,000/231 = 623.38).

Input: Length and Width of roof in ft, inches of rainfall
Output: Total runoff in in^3
Process: (((roof_width x roof_length)*12)*in_rain)/231

BEGIN

    prompt user for roof_width
    get roof_width
    prompt user for roof_length
    get roof_length
    Prompt User for in_rain
    get in_rain
     
    roof_width = roof_width * 12
    roof_length = roof_length * 12
    
    
    cub_in_rain = roof_width * roof_length * in_rain
    gallon = cub_in_rain/231
    display gallon
    
END
    
Test:
roof_width = 10
roof_length = 10
in_rain = 10

roof_width = 10 * 12 = 120
roof_length = 10 * 12 = 120

cub_in_rain = 120 * 120 * 10 = 144000
gallon = 144,000/231
gallon = 623.38

#### 4. Calculate employee's pay

    Endif
    
    straight_pay = straight_time * pay_rate
    weekly_pay = straight_pay + overtime_pay
